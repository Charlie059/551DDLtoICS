```python
FileasList = file.readlines()
```

```python
from pandas.core.frame import DataFrame
import pandas as pd
import datetime
timeTable = pd.DataFrame(columns = ['Subject' , 'Start Date', 'Start Time'])
```

```python
from pandas.core.indexes.datetimes import date_range
import time
for i in range(len(FileasList)):
  if(i%3 == 0):
    # print(FileasList[i])
    j = 0
    date = ""
    d= datetime
    t=time
    ddLName = ""
    for name in FileasList[i].split():   
      if(j == 0):
        print(name) 
        ddLName = name
      if(j == 3): # MM
        date = date + name + "\\"
      if(j == 4): # DD
        date = date + name + "\\"
      if(j == 6): # YY
        date = date + name
        d = datetime.datetime.strptime(date, '%b\%d\%Y') # Date
        
      if(j == 5): # Time
        t = time.strftime('%I:%M %p', time.strptime(name, '%H:%M:%S')) # Time
      j = j + 1
    timeTable.loc[int(i/3)] = (ddLName, d, t)
# 05/30/2020 10:00 AM
```
